#### Imports

```python
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

#### Target Variable

```python
# Describe
y_train = data.loc[:,'label']
display(y_train)
```

#### Explanatory Variables

```python
x_train = data.loc[:,data.columns[1:]]
display(x_train.head())
```

```python
# Scale x_train using Standard Scaling

scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = pd.DataFrame(scaler.transform(x_train))
display(x_train_scaled.head())
```

#### Create Training and Validation Sets

```python
x_train_cv, x_valid, y_train_cv, y_valid = \
train_test_split(x_train_scaled, y_train, test_size = 0.2, train_size = 0.8)

print('x_train_cv')
display(x_train_cv.head())

print('x_valid')
display(x_valid.head())

print('y_train_cv')
display(y_train_cv.head())

print('y_valid')
display(y_valid.head())
```

#### Test Set

```python
standard_scaler = StandardScaler()
standard_scaler.fit(test)
test_scaled = pd.DataFrame(standard_scaler.transform(test))
display(test_scaled.head())
```
